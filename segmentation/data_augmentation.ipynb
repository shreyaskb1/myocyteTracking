```python
import numpy as np
from matplotlib import pyplot as plt
from skimage.transform import AffineTransform, warp
from skimage import io, img_as_ubyte
import random
import os
from scipy.ndimage import rotate
import cv2
SIZE_X = 512
SIZE_Y = 512

import albumentations as A
images_to_generate = 2000
```

```python
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
```

```python
images_path="/content/gdrive/MyDrive/REUProject2021/images_short_axis/images_clr/" 
masks_path = "/content/gdrive/MyDrive/REUProject2021/images_short_axis/masks_bw/"
img_augmented_path="gdrive/MyDrive/REUProject2021/images_short_axis/augmented_images" 
msk_augmented_path="gdrive/MyDrive/REUProject2021/images_short_axis/augmented_masks" 
images=[] 
masks=[]
```

```python
for im in os.listdir(images_path):  # read image name from folder and append its path into "images" array     
    images.append(os.path.join(images_path,im))

for msk in os.listdir(masks_path):  # read image name from folder and append its path into "images" array     
    masks.append(os.path.join(masks_path,msk))

images = sorted(images)
masks = sorted(masks)
```

```python
print(sorted(images))
print(sorted(masks))
```

```python
aug = A.Compose([
    A.VerticalFlip(p=0.5),              
    A.RandomRotate90(p=0.5),
    A.HorizontalFlip(p=0.8),
    A.Transpose(p=0.7),
    A.GridDistortion(p=0.5),
    A.RandomBrightnessContrast(p=0.3)
    ]
)


i = 1  


while i<=images_to_generate: 
    number = random.randint(0, len(images)-1)  
    image = images[number]
    mask = masks[number]
    print(i, image, mask)

    original_image = io.imread(image)
    original_mask = io.imread(mask)
    
    augmented = aug(image=original_image, mask=original_mask)
    transformed_image = augmented['image']
    transformed_mask = augmented['mask']
    
    transformed_image = cv2.resize(transformed_image, (SIZE_Y, SIZE_X))
    transformed_image = cv2.cvtColor(transformed_image, cv2.COLOR_RGB2BGR)

    transformed_mask = cv2.resize(transformed_mask, (SIZE_Y, SIZE_X))
    
        
    new_image_path= "%s/augmented_image_%s.png" %(img_augmented_path, i)
    new_mask_path = "%s/augmented_mask_%s.png" %(msk_augmented_path, i)

    io.imsave(new_image_path, transformed_image)
    io.imsave(new_mask_path, transformed_mask)
    i = i + 1
```
